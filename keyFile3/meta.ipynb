```python
import pandas as pd
#importing yf library , yahoo finance
import yfinance as yf  
import matplotlib.pyplot as plt
```

```python
meta = yf.download(["META","PINS"])
#download historical data
```

```python
meta.index = pd.to_datetime(meta.index)
# converting to datime objects
```

```python
data = meta.loc["2013":]
```

```python
#data plotting, creating secondary y axis

fig , ax1 = plt.subplots(layout="constrained")
# plotting meta and pins 
ax1.plot(data["Close"]["META"]) 
# ax2 = ax1.twinx()
ax2 = ax1.twinx()
ax2.plot(data["Close"]["PINS"],color="orange")
# adding labels to left y axis 
ax1.set_ylabel("PINS STOCK PRICES (in Orange)")
# adding labels to right y axis , adding title 
ax2.set_ylabel("META STOCK PRICES (in Blue)")
plt.title('Stock Prices of META and PINS')

# adding grid 
plt.grid()
plt.show();
```

```python
data["Close"]["META"].plot(secondary_y=True)
data["Close"]["PINS"].plot()


#data plotting, creating secondary y axis
```
